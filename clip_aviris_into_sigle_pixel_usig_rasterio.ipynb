```python
from osgeo import gdal,ogr,osr
import numpy as np
from PIL import Image
import math
import rasterio

from rasterio.plot import show

from rasterio.plot import show_hist

from rasterio.mask import mask
from affine import Affine

from shapely.geometry import box
import geopandas as gpd
from fiona.crs import from_epsg
from rasterio.windows import Window
# import pycrs
```

```python
# open the source raster file
with rasterio.open(r'F:\wenqu\Aviris\site1b\clipped_aviris\site1b_aviris_clip') as src:
    # height is number of rows and width is the number of columns, x-22, y-26
    print(src.height, src.width)
    for y in range(src.height):
        for x in range(src.width):
            print(x, y)
            window = Window(x,y,1,1)
            # Get the transform of the parent dataset
            parent_transform = src.transform

            # Get the offset of the window
            col_off, row_off = window.col_off, window.row_off

            # Define the transform of the window
            window_transform = parent_transform * Affine.translation(col_off, row_off)
            
            # create a new raster file for the pixel values
            metadata = src.meta.copy()
            print(metadata)
            metadata.update({
                "height": 1,
                "width": 1,
                "count": 425,  # set count to the number of bands in the source raster
                "transform": window_transform
            })
            output = "F:/wenqu/Aviris/site2a/PLSR/aviris_single_pixel/" + str(x) + '_' + str(y) + '.tif'
            with rasterio.open(output, "w", **metadata) as dst:
                # loop over the bands and write the pixel value for each band
                for band in range(1, src.count + 1):
                    src_data = src.read(band, window=window)
                    dst.write(src_data, indexes=band)

```
