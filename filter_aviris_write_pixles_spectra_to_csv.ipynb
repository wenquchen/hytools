```python
from osgeo import gdal
import os
import shutil
import csv
```

```python
# Define the directory path
folder_path = "D:/site2a/PLSR/site2a_clip"
keep_file = []
na_files = []
```

```python
trait_file_list = []
for filename in os.listdir(folder_path):
    if filename.endswith('.tif'):
        x = filename.split(".")
        trait_file_list.append(x[0])
#         print(x[0])
```

```python
len(trait_file_list)
```

```python
source_folder = r'D:/site2a/PLSR/site2a_aviris_clip/'
destination_folder = r'D:/site2a/PLSR/corresponnding_aviris/'
```

```python
csv_path = "D:/site2a/PLSR/output.csv"

# Create a CSV file
csv_file = open(csv_path, "w", newline='')
writer = csv.writer(csv_file)
```

```python
aviris_bands = 122
```

```python
import csv
with open('D:/site2a/PLSR/output1.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['file_name'] + [f'band_{band}' for band in range(1, aviris_bands + 1)])
    
    for f_name in trait_file_list:
        csv_row = [f_name]
    
        dataset = gdal.Open(os.path.join(destination_folder, f_name))
        num_bands = dataset.RasterCount 
    
    
    
    
        for band in range(1, num_bands+1):
            # Get the band object
            band_obj = dataset.GetRasterBand(band)
            # Read the band data
            csv_row.append(band_obj.ReadAsArray())
        writer.writerow(csv_row)
```

```python
hdr_file = []
for file in trait_file_list:
    hdr = file + '.hdr'
    hdr_file.append(hdr)
#     print(hdr)
```

```python
# fetch all files
for file_name in trait_file_list:
    # construct full file path
    source = source_folder + file_name
    destination = destination_folder + file_name
    # copy only files
    if os.path.isfile(source):
        shutil.copy(source, destination)
        print('copied', file_name)
```
