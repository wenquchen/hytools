```python
import pandas as pd
import numpy as np
import csv
import geopandas as gpd
import geopandas
from pyproj import Proj, transform
from shapely.geometry import Point
```

```python
df = pd.read_csv(r"F:\wenqu\GPS\2022_GPS\GPS_CSV\siteb_processed_update1\quadrat.csv", skip_blank_lines=True)
```

```python
df
```

```python
name_list = df['Name'].tolist()
lat_list = df['Latitude'].tolist()
lon_list = df['Longitude'].tolist()
```

```python
inProj = Proj(init='epsg:4326')
outProj = Proj(init='epsg:32605')
```

```python
x_cor = []
y_cor = []
name_attribute_list = []
for i in range(len(name_list)):

#     print(i)    
    x, y = transform(inProj,outProj, lon_list[i],lat_list[i])        
    x_cor.append(x)
    y_cor.append(y)
    name_attribute_list.append(name_list[i])
```

```python
x_center = []
y_center = []
center_attribute_list = []
for i in range(int(len(name_list)/2)):
    print(i)    
    x1, y1 = transform(inProj,outProj, lon_list[2*i],lat_list[2*i])
    x2, y2 = transform(inProj,outProj, lon_list[2*i + 1],lat_list[2*i + 1])
    x_c = (x1 + x2)/2
    y_c = (y1 + y2)/ 2

        
    x_center.append(x_c)
    y_center.append(y_c)
    center_attribute_list.append(name_list[2*i])
```

```python
df_center = pd.DataFrame(
    {'f_name': center_attribute_list,
     'X': x_center,
     'Y': y_center})
```

```python
df_center
```

```python
inProj_back = Proj(init='epsg:32605')
outProj_back = Proj(init='epsg:4326')
```

```python
df_center['f_name'][0]
```

```python
x_cor_back = []
y_cor_back = []
# name_attribute_list = []
for i in range(len(df_center['f_name'])):

#     print(i)    
    x, y = transform(inProj_back,outProj_back, df_center['X'][i],df_center['Y'][i])        
    x_cor_back.append(x)
    y_cor_back.append(y)
#     name_attribute_list.append(name_list[i])
```

```python
df_center_latlong = pd.DataFrame(
    {'f_name': center_attribute_list,
     'X': x_cor_back,
     'Y': y_cor_back})
```

```python
df_center_latlong
```

```python
geometry = [Point(xy) for xy in zip(df_center_latlong['X'], df_center_latlong['Y'])]
gdf = gpd.GeoDataFrame(df_center_latlong, crs='EPSG:4326', geometry=geometry)
gdf.to_file(r'F:\wenqu\GPS\shp\site1b\update1\site1b_quadrant_center_update1.shp', driver='ESRI Shapefile')
```
