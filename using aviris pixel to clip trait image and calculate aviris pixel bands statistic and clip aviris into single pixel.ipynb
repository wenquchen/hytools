```python
from osgeo import gdal,ogr,osr
```

```python
aviris_img_org = gdal.Open( 'F:/wenqu/Aviris/site2a/resample/aviris_resample_clip_final_update3')
aviris_org_geotransform = aviris_img_org.GetGeoTransform()
aviris_org_geotransform
```

```python
aviris_img_array = aviris_img_org.ReadAsArray()
aviris_img_array.shape
```

```python
aviris_bands = aviris_img_org.RasterCount
aviris_driver = aviris_img_org.GetDriver()
aviris_projection = aviris_img_org.GetProjection()
aviris_geotransform = aviris_img_org.GetGeoTransform()
```

```python
# Loop through each band of the image
# Create a list to store the pixel values
pixel_values = []

for band in range(1, aviris_bands + 1):

    # Get the band object and read the pixel values into a list
    band_obj = aviris_img_org.GetRasterBand(band)
    band_data = band_obj.ReadAsArray().tolist()

    # Append the pixel values to the pixel_values list
    pixel_values.append(band_data)
```

```python
# Write the pixel values to a CSV file
import csv
with open('F:/wenqu/trait_map/aviris_pixel_data1.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['row', 'column'] + [f'band_{band}' for band in range(1, aviris_bands + 1)])
    # Write the pixel values to the CSV file
    for row in range(aviris_img_org.RasterYSize):
        for col in range(aviris_img_org.RasterXSize):
            csv_row = [row , col ]
            for band in range(aviris_bands):
                csv_row.append(pixel_values[band][row][col])
            writer.writerow(csv_row)
```
