```python
import os
from osgeo import gdal,ogr,osr
import csv
```

```python
img_path = r'D:\wenqu\chapter1_2\updated_aviris\PLSR_training_samples\AVIRIS_NG_Spectra'
delimiter = "_site1b"
aviris_file = []
# Get all file names in the folder
file_names = [f for f in os.listdir(img_path) if os.path.isfile(os.path.join(img_path, f))]
# Extract the part of file names after the delimiter
extracted_names = [file_name.split(delimiter, 1)[1] for file_name in file_names if delimiter in file_name]

# Print the extracted names
for extracted_name in extracted_names:
    f_name = 'site1b' + extracted_name
    aviris_file.append(f_name)
```

```python
aviris_bands = 425
destination_folder = 'F:/wenqu/Aviris/site1b/aviris_single_pixel1/'
```

```python
import csv
with open('F:/wenqu/Aviris/site1b_aviris.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['file_name'] + [f'band_{band}' for band in range(1, aviris_bands + 1)])
    
    for f_name in aviris_file:
        print(f_name)
        csv_row = [f_name]
    
        dataset = gdal.Open(os.path.join(destination_folder, f_name))
        num_bands = dataset.RasterCount 
    
    
    
    
        for band in range(1, num_bands+1):
            # Get the band object
            band_obj = dataset.GetRasterBand(band)
            # Read the band data
            csv_row.append(band_obj.ReadAsArray())
        writer.writerow(csv_row)
```
