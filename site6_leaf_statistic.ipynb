```python
import pandas as pd
```

```python
# read specific columns of csv file using Pandas
df_plant_cover = pd.read_csv(r"F:\wenqu\DOE_Arctic_project\2021_data\leafwtAK2021_site6A.csv", usecols = ['Site','quadrat', 'spp', 'cover', 'sla', 'la', 'LDMC']).dropna()
df_plant_cover
```

```python
df_plant_cover["site_quadrant"] = df_plant_cover['Site'].astype(str) + '_' + df_plant_cover['quadrat'].astype(int).astype(str)
df_plant_cover["site_quadrant_species"] = df_plant_cover['Site'].astype(str) + '_' + df_plant_cover['quadrat'].astype(int).astype(str) + '_' + df_plant_cover["spp"]
df_plant_cover
```

```python
df1 = df_plant_cover.groupby(['Site','quadrat'], as_index=False)['cover'].sum()
df1.rename(columns={'cover': 'sum_cover'}, inplace=True)
df1["site_quadrant"] = df1['Site'].astype(str) + '_' + df1['quadrat'].astype(int).astype(str)
df1
```

```python
df_combine = pd.merge(df_plant_cover, df1, on ='site_quadrant')
df_combine
```

```python
df = df_combine[['site_quadrant', 'site_quadrant_species', 'sla', 'la', 'LDMC', 'cover', 'sum_cover']]

df['percent_cover'] = df['cover'] / df['sum_cover']
df
```

```python
df['weighted_sla'] = df['sla'].astype(float) * df['percent_cover'].astype(float)
df['weighted_la'] = df['la'].astype(float) * df['percent_cover'].astype(float)
df['weighted_LDMC'] = df['LDMC'].astype(float) * df['percent_cover'].astype(float)
df
```

```python
df_final = df.groupby(['site_quadrant']).agg({'weighted_sla':'sum','weighted_la':'sum','weighted_LDMC':'sum'})
df_final
```

```python
df_final.to_csv(r'F:\wenqu\DOE_Arctic_project\2021_statistic\site6_weighted_leaf.csv') 
```
