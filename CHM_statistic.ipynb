```python
path = 'F:/wenqu/DEM/site1b/clip'
hdr_files = []
for f in os.listdir(path):
    if f.endswith('.tif'):
        hdr_files.append(f.replace('.tif', ''))
```

```python
file_band_mean = []
file_band_std = []
for i in range(len(hdr_files)):
    dst_hdr = gdal.Open('F:/wenqu/DEM/site1b/clip/'+ hdr_files[i]+'.tif').ReadAsArray() 
    
    band_mean = dst_hdr.mean()
    file_band_mean.append(band_mean)
    band_std = dst_hdr.std()
    file_band_std.append(band_std)
```

```python
pxy = {'site': hdr_files, 
     'CHM_mean': file_band_mean,
     'CHM_std': file_band_std
    }
```

```python
CHM_df = pd.DataFrame.from_dict(pxy)
CHM_df
```

```python
CHM_df.to_csv(r'F:\wenqu\DEM\site1b\site1b_CHM.csv')
```
