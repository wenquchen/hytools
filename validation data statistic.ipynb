```python
# specify the folder path
folder_path = 'F:/wenqu/trait_map/validation_dataset/d13c_validation'

# get all entries in the folder using listdir
all_entries = os.listdir(folder_path)

# filter out only files from the entries (ignoring subdirectories)
file_names = [entry for entry in all_entries if os.path.isfile(os.path.join(folder_path, entry))]

print(file_names)
```

```python
# folder_path = 'F:/wenqu/trait_map/validation_dataset/validation_csv'
# file_names = [entry for entry in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, entry))]

# Open a CSV file for writing
with open('F:/wenqu/trait_map/validation_dataset/validation_csv/d13c.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Write the header row
    csvwriter.writerow(['File Name', 'Ymax1', 'Std_New1'])

    for i in file_names:
        file_data = gdal.Open(os.path.join(folder_path, i)).ReadAsArray().flatten()

        kde1 = gaussian_kde(file_data)
        x1 = np.linspace(np.min(file_data), np.max(file_data), 1000)
        pdf1 = kde1(x1)
        ymax1 = np.max(pdf1)

        new_data1 = kde1.resample(40000)[0]
        std_new1 = np.std(new_data1)

        print(f'File: {i}, Std_New1: {std_new1}')

        # Write the results to the CSV
        csvwriter.writerow([i, ymax1, std_new1])

print("Data saved to output.csv")
```
