```python
import rasterio
from rasterio.mask import mask
from rasterio import Affine
import geopandas as gpd
from shapely.geometry import box
from rasterio.features import geometry_mask
import numpy as np
from shapely.geometry import mapping
from shapely.geometry import Polygon
import os
from osgeo import gdal,ogr,osr
import matplotlib.pyplot as plt
```

```python
del_file_name = []
keep_file_name = []
```

```python
import os
from osgeo import gdal,ogr,osr
folder_path = "F:/wenqu/Aviris/site1b/cliped_trait"
extension = ".tif"

for filename in os.listdir(folder_path):
    # Check if the current item is a file
    if filename.endswith(extension) and os.path.isfile(os.path.join(folder_path, filename)):
        print(filename)
        x_data = gdal.Open(os.path.join(folder_path, filename)).ReadAsArray()
        x_data = x_data[x_data>0]
        if x_data.shape[0] >22000:
            keep_file_name.append(filename)
        else:
            del_file_name.append(filename)
            
```

```python
print(del_file_name)
```

```python
len(del_file_name)
```

```python
# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    # Check if the file has the desired extension
    if filename in del_file_name:
        # Create the file path and remove the file
        file_path = os.path.join(folder_path, filename)
        os.remove(file_path)
```

```python
aviris_keep = []
aviris_del = []
```

```python
import os
from osgeo import gdal,ogr,osr
aviris_folder_path = "F:/wenqu/Aviris/site1b/single_aviris_pixel"
extension = ".tif"

for filename in os.listdir(aviris_folder_path):
    # Check if the current item is a file
    if filename.endswith(extension):
        print(filename)
        if filename in keep_file_name:
            aviris_keep.append(filename)
        else:
            aviris_del.append(filename)
```
