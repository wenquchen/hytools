```python
import numpy as np
import pandas as pd
from osgeo import gdal
import os
import csv
```

```python
import os
file_name = []
folder_path1 = r"F:\wenqu\trait_map\trait_clip\site6_d15n"
site1b_delimiter = "_site6"

for filename1 in os.listdir(folder_path1):
    i = 'site6' + filename1.split(site1b_delimiter, 1)[1]
    print(i)
    file_name.append(i)

```

```python
len(file_name)
```

```python
# path to the directory containing the TIFF files
tiff_dir = r'F:\wenqu\Aviris\site6\site6_aviris_single_pixel1'
tiff_files  = []
for aviris_file in os.listdir(tiff_dir):
    if aviris_file in file_name:
        tiff_files.append(aviris_file)
```

```python
len(tiff_files)
```

```python
band_data = []

for f_name in tiff_files:
    print(f_name)
    csv_row = [f_name]

    dataset = gdal.Open(os.path.join(tiff_dir, f_name))
    num_bands = dataset.RasterCount 
    for band in range(1, num_bands+1):
        # Get the band object
        band_obj = dataset.GetRasterBand(band)
        # Read the band data
        csv_row.append(band_obj.ReadAsArray())
    band_data.append(csv_row)
#         writer.writerow(csv_row)
```

```python

with open('F:/wenqu/Aviris/Aviris_csv/csv/site6_aviris_single_pixel_band.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    headers = ['File Name'] + [f'band_{i}' for i in range(1, 426)]
    writer.writerow(headers)
    for row in band_data:
        writer.writerow(row)
```
