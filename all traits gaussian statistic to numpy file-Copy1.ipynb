# d13c

```python
origin_ymax = []
origin_std = []
resampled_ymax = []
resampled_std = []

input_folder = r'D:\wenqu\chapter1_2\updated_uas_modeling\plsr_training\version4\pn'
# output_folder = r"E:\wenqu\probability_density_function\without_dem_numpy\validation"

for i in sorted_site_name:
    file_name = i
    print(i)
    
    site_array = gdal.Open(os.path.join(input_folder, file_name)).ReadAsArray()

    # Clean NaNs and remove background (0)
    data_clean = site_array[np.isfinite(site_array)]           # Remove NaN or inf
    data_clean = data_clean[data_clean != 0]                   # Remove background (0)

    # Check if there are still values left after cleaning
    if data_clean.size == 0:
        print(f"Skipped {file_name}: only background or invalid values.")
        continue  # Skip to next file if no valid data

    site_array = data_clean.flatten()
    print(site_array.shape)

    # Original KDE
    kde1 = gaussian_kde(site_array)
    x1 = np.linspace(np.min(site_array), np.max(site_array), 1000)
    pdf1 = kde1(x1)

    n_samples = len(site_array)
    new_data = kde1.resample(n_samples).flatten()

    x_new = np.linspace(np.min(new_data), np.max(new_data), 1000)
    x1_max_density = x1[np.argmax(pdf1)]
    origin_ymax.append((x1_max_density, file_name))
    x1_std_value = np.std(x1)
    origin_std.append((x1_std_value, file_name))

    # Resampled KDE
    kde_new_data = gaussian_kde(new_data)
    pdf_new_data = kde_new_data(x_new)
    resampled_max_density = x_new[np.argmax(pdf_new_data)]
    
    resampled_ymax.append((resampled_max_density, file_name))

    resampled_std_value = np.std(new_data)
    resampled_std.append((resampled_std_value, file_name))



```

```python


# combined_origin_ymax = np.array(origin_ymax, dtype=object)
# np.save('E:/wenqu/trait_map/update4/training/gaussian_numpy/origin_ymax_d13c.npy', combined_origin_ymax)

# combined_origin_std = np.array(origin_std, dtype=object)
# np.save('E:/wenqu/trait_map/update4/training/gaussian_numpy/origin_std_d13c.npy', combined_origin_std)

combined_resampled_ymax = np.array(resampled_ymax, dtype=object)
np.save(r'D:\wenqu\chapter1_2\updated_uas_modeling\plsr_training\version4\numpy\ymax_pn.npy', combined_resampled_ymax)

combined_resampled_std = np.array(resampled_std, dtype=object)
np.save(r'D:\wenqu\chapter1_2\updated_uas_modeling\plsr_training\version4\numpy\std_pn.npy', combined_resampled_std)
```
