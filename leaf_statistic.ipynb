```python
import pandas as pd
```

```python
# read specific columns of csv file using Pandas
df_plant_cover = pd.read_csv(r"F:\wenqu\DOE_Arctic_project\2022_data\Summer2022_abundance.csv", usecols = ['Site','Quadrat', 'species', 'abundance_per']).dropna()
df_plant_cover
```

```python
df_plant_cover["site_quadrant_species"] = df_plant_cover['Site'].astype(str) + '_' + df_plant_cover['Quadrat'].astype(int).astype(str) + '_' + df_plant_cover["species"]
df_plant_cover
```

```python
df_leaf = pd.read_csv(r"F:\wenqu\DOE_Arctic_project\2022_data\Summer2022_leaf.csv", usecols = ['Site','Quadrat', 'species', 'sla', 'la', 'ldmc']).dropna()
df_leaf
```

```python
df_leaf["site_quadrant_species"] = df_leaf['Site'].astype(str) + '_' + df_leaf['Quadrat'].astype(int).astype(str) + '_' + df_leaf["species"]
df_leaf
```

```python
df_combine = pd.merge(df_leaf, df_plant_cover, on ='site_quadrant_species')
df_combine
```

```python
df_combine["site_quadrant"] = df_combine['Site_x'].astype(str) + '_' + df_combine['Quadrat_x'].astype(int).astype(str) 
df_combine
```

```python
df1 = df_combine.groupby(['Site_x','Quadrat_x'], as_index=False)['abundance_per'].sum()
df1["site_quadrant"] = df1['Site_x'].astype(str) + '_' + df1['Quadrat_x'].astype(int).astype(str) 
df1.rename(columns={'abundance_per': 'sum_cover'}, inplace=True)
df1
```

```python
new_df = pd.merge(df_combine, df1,  how='left', left_on=['site_quadrant'], right_on = ['site_quadrant'])
new_df
```

```python
df = new_df[['site_quadrant', 'site_quadrant_species', 'sla', 'la', 'ldmc', 'abundance_per', 'sum_cover']]

df['percent_cover'] = df['abundance_per'] / df['sum_cover']
df
```

```python
df['weighted_sla'] = df['sla'].astype(float) * df['percent_cover'].astype(float)
df['weighted_la'] = df['la'].astype(float) * df['percent_cover'].astype(float)
df['weighted_LDMC'] = df['ldmc'].astype(float) * df['percent_cover'].astype(float)
df
```

```python
df_final = df.groupby(['site_quadrant']).agg({'weighted_sla':'sum','weighted_la':'sum','weighted_LDMC':'sum'})
df_final
```

```python
df_final.to_csv(r'F:\wenqu\DOE_Arctic_project\2022_data_statistics\site1b_weighted_leaf.csv') 
```
