```python
import shutil
import os



# Define your source and destination directories
source_dir = r'E:\wenqu\trait_map\trait_map_without_chm\clip\site6_pn'
destination_dir = r'E:\wenqu\trait_map\trait_map_without_chm\training\new\pn'

# Make sure the destination directory exists, if not, create it
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Iterate over the filenames and copy each one to the destination directory if it exists in the source directory
for filename in filenames:
    source_path = os.path.join(source_dir, filename)
    destination_path = os.path.join(destination_dir, filename)
    
    # Check if the file exists before copying
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_path)
    else:
        print(f"File {filename} does not exist in the source directory. Skipping.")

```
