```python
def modify_filename_with_prefix(filename, prefix="sla_"):
    # Use regex to find the pattern of numbers
    match = re.search(r'_(\d+)_(\d+)', filename)
    if match:
        # Keep the first part (e.g., site6) unchanged and convert only the last two numbers to three-digit format
        first_num = match.group(1).zfill(3)
        second_num = match.group(2).zfill(3)
        # Replace the matched part with the modified numbers
        modified_name = prefix + filename.replace(f"_{match.group(1)}_{match.group(2)}", f"_{first_num}_{second_num}")
        return modified_name
    return prefix + filename  # Add prefix even if no match found

# Update the filenames in the array
updated_array = d13c_trait_pdf_ymax.copy()
updated_array[:, 1] = [modify_filename_with_prefix(filename) for filename in updated_array[:, 1]]

# Save the updated array
np.save(r"D:\wenqu\chapter1_2\updated_aviris\PLSR_training_samples\trait_numpy\updated_ymax_sla.npy", updated_array)

# Print the updated array to verify
print(updated_array)  # Display the first 5 rows
```

```python
# Function to standardize the filename with leading zeros
def standardize_filename(filename):
    parts = filename.split("_")
    
#     # Keep only the digits from the site name and add leading zeros where necessary
#     site_name = parts[1]
#     parts[1] = "site" + ''.join(filter(str.isdigit, site_name))
    
    # Format the last two parts to have leading zeros
    parts[2] = f"{int(parts[2]):02d}"
    parts[3] = f"{int(parts[3].split('.')[0]):03d}.tif"
    new_name = parts[0] + '_' + parts[1] +'_' + parts[2] + '_' + parts[3]
    
    return new_name
```

```python
d13c_trait_pdf_ymax_sorted = np.copy(d13c_trait_pdf_ymax)
# Apply the function to the array
for i in range(d13c_trait_pdf_ymax.shape[0]):
    d13c_trait_pdf_ymax_sorted[i, 1] = standardize_filename(d13c_trait_pdf_ymax[i, 1])

print(d13c_trait_pdf_ymax_sorted)
```

```python
# Getting the indices that would sort the 'sla_site...' column
sorted_indices = np.argsort(d13c_trait_pdf_ymax_sorted[:, 1])

# Sorting the array
sorted_data = d13c_trait_pdf_ymax_sorted[sorted_indices]
print(sorted_data)
# Saving the sorted array as .npy file
np.save('E:/wenqu/probability_density_function/without_dem_numpy/sorted/resampled_ymax_d13c.npy', sorted_data)
```

```python
import os

# Define the directory path
directory_path = r'E:\wenqu\trait_map\trait_map_without_chm\clip\copy\combine\d13c'

# Iterate over each file in the directory
for filename in os.listdir(directory_path):
    # Skip directories
    if os.path.isdir(os.path.join(directory_path, filename)):
        continue
    
    # Split the filename into parts
    parts = filename.split("_")
    
    # Ensure there are at least four parts (including the file extension part)
    if len(parts) != 4:
        print(f"Skipping unexpected filename format: {filename}")
        continue
    
    # Extract the file extension
    file_stem, parts_end = parts[3].split(".")
    
    # Pad the parts with zeros
    part2_padded = parts[2].zfill(2)
    part3_padded = file_stem.zfill(3)
    
    # Construct the new filename
    new_filename = f"{parts[0]}_{parts[1]}_{part2_padded}_{part3_padded}.{parts_end}"
    
    # Print the details
    print(parts, part2_padded, part3_padded, parts_end, new_filename)
    
    # Rename the file
    old_file_path = os.path.join(directory_path, filename)
    new_file_path = os.path.join(directory_path, new_filename)
    
    # Check if the new filename already exists to avoid overwriting
    if os.path.exists(new_file_path):
        print(f"Skipping rename, file already exists: {new_filename}")
        continue
    
    # Rename the file
    os.rename(old_file_path, new_file_path)
    print(f"Renamed {filename} to {new_filename}")
```
