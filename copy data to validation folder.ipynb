```python
# Define the folder path
folder_path = r'E:\wenqu\trait_map\update1\trait_map_with_scale\validation\validation_add6\d13c'  # Replace with your folder path

# Search for all .tif files in the folder
tif_files = glob.glob(os.path.join(folder_path, '*.tif'))

# Extract just the filenames
file_names = [os.path.basename(file) for file in tif_files]


copy_file = []
# Now, file_names contains only the names of the .tif files
for name in file_names:
    filename = os.path.basename(name)
    filename_without_extension, _ = os.path.splitext(filename)
    parts = filename_without_extension.split('_')
    
    shp_name = parts[1] + '_' + parts[2] + '_' + parts[3] + '.shp'
    copy_file.append(shp_name)
    dbf_name = parts[1] + '_' + parts[2] + '_' + parts[3] + '.dbf'
    copy_file.append(dbf_name)
    prj_name = parts[1] + '_' + parts[2] + '_' + parts[3] + '.prj'
    copy_file.append(prj_name)
    shx_name = parts[1] + '_' + parts[2] + '_' + parts[3] + '.shx'
    copy_file.append(shx_name)
    print(name, parts, shp_name)
```
