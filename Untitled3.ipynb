```python
import csv

# Function to process each line by keeping site, two coordinates, and modifying the filename
def process_line(line):
    # Split the line into parts
    parts = line.strip().split()

    # Debugging: print the parts after splitting the line
    print(f"Split parts: {parts}")
    
    if len(parts) >= 5:
        # The second column is the site
        site = parts[1]
        
        # The third and fourth columns are coordinates, removing the brackets
        lat = parts[2].strip('[]')
        lon = parts[3].strip('[]')
        
        # The last part is the filename, we remove the '.tif'
        filename = parts[4].strip("'")[:-4]  # Remove the last 4 characters (.tif)

        # Debugging: print what will be written to CSV
        print(f"Processed line: {site}, {lat}, {lon}, {filename}")
        
        return [site, lat, lon, filename]

    return []  # Return empty if the line doesn't match the expected format
# Read the file
input_file = r'E:\wenqu\trait_map\file_name.txt'  # Replace with your input file path
output_file = r'E:\wenqu\trait_map\pn.csv'  # Replace with your output CSV file path

# Read the input file and write to the output CSV
with open(input_file, 'r') as infile, open(output_file, 'w', newline='') as outfile:
    csv_writer = csv.writer(outfile)
    
    for line in infile:
        processed_line = process_line(line)
        if processed_line:
            csv_writer.writerow(processed_line)  # Write the processed line to CSV

print(f"Processed file saved to {output_file}")
```

```python
import pandas as pd

# File path to your text file
input_file = r'D:\wenqu\chapter1_2\figures\aviris_d13c_ymax.txt'  # Replace with the correct file path

# Try reading the file with a different approach to handle irregular data
with open(input_file, 'r') as file:
    lines = file.readlines()

# Print the first few lines to check if the file is read properly
print("First few lines from the file:")
print(lines[:5])

# Now manually process each line and remove unnecessary characters
processed_lines = []
for line in lines:
    parts = line.strip().split()  # Split by whitespace
    
    # Only process lines with enough parts (e.g., 5 parts: index, site, lat, lon, filename)
    if len(parts) >= 5:
        site = parts[1]
        lat = parts[2].strip('[]')
        lon = parts[3].strip('[]')
        filename = parts[4].strip("'").replace(".tif", "")
        
        # Append the processed data
        processed_lines.append([site, lat, lon, filename])

# Convert the processed lines into a DataFrame
df = pd.DataFrame(processed_lines, columns=['Site', 'Latitude', 'Longitude', 'FileName'])

# Display the DataFrame to ensure it's correct
print(df.head())

# Save to CSV
df.to_csv('output.csv', index=False)
print("CSV file saved as 'output.csv'.")
```
