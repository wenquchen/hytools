```python
import os
import random
import rasterio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Folder where your .tif files are stored
folder_path = r'E:\wenqu\trait_map\trait_map_without_chm\training\new_Copy\la'

# Site names
sites = ['site2a', 'site2c', 'site6', 'site1b']

# Number of random samples
num_samples = 100

# Initialize a dictionary to store the summaries for each site
site_data = {}

# Function to extract data from .tif file
def summarize_tif(file_path):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        return np.nanmean(data), np.nanmedian(data), np.nanstd(data)  # Example summary statistics

# Loop through each site
for site in sites:
    # List all files for the current site
    site_files = [f for f in os.listdir(folder_path) if site in f and f.endswith('.tif')]
    
    # Randomly select 100 files
    selected_files = random.sample(site_files, min(num_samples, len(site_files)))
    
    # Initialize list to store summaries
    summaries = []
    
    # Process each selected file
    for file in selected_files:
        file_path = os.path.join(folder_path, file)
        mean_val, median_val, std_val = summarize_tif(file_path)
        summaries.append({'mean': mean_val, 'median': median_val, 'std': std_val})
    
    # Store the summaries for the current site
    site_data[site] = pd.DataFrame(summaries)

# Create box plots for each site
for site, data in site_data.items():
    plt.figure()
    data.boxplot(column=['mean', 'median', 'std'])
    plt.title(f'Summary Statistics for {site}')
    plt.ylabel('Values')
    plt.show()
```

```python
import os
import random
import rasterio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Folder where your .tif files are stored
folder_path = r'E:\wenqu\trait_map\trait_map_without_chm\training\new_Copy\la'

# Site names
sites = ['site1b','site2a', 'site2c', 'site6' ]

# Number of random samples
num_samples = 100
random.seed(42)  # You can change 42 to any number to set a different seed
# Initialize a dictionary to store the mean values for each site
site_data = {}
# Initialize a dictionary to store the mean values for each site
site_data = {}

# Function to extract data from .tif file and calculate mean (excluding 0 values)
def calculate_mean_tif(file_path):
    with rasterio.open(file_path) as src:
        data = src.read(1)  # Read the first band
        # Mask out 0 values (background) before calculating the mean
        data = np.where(data == 0, np.nan, data)  # Replace 0 with NaN
        return np.nanmean(data)  # Return mean, ignoring NaN values

# Loop through each site
for site in sites:
    # List all files for the current site
    site_files = [f for f in os.listdir(folder_path) if site in f and f.endswith('.tif')]
    
    # Randomly select 100 files
    selected_files = random.sample(site_files, min(num_samples, len(site_files)))
    
    # Calculate the mean for each selected file and store in a list
    means = [calculate_mean_tif(os.path.join(folder_path, file)) for file in selected_files]
    
    # Store the list of mean values for the current site
    site_data[site] = means

# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(site_data.values(), labels=site_data.keys())
plt.xlabel('Site')
plt.ylabel('Mean Value (Excluding 0 Background)')
plt.title('Distribution of Mean Values Across Sites (Excluding Background)')
plt.show()
```
