```python
import pandas as pd
```

```python


# Paths to your CSV files
csv_file1 = r'F:\wenqu\DOE_Arctic_project\2021_statistic\leafwtAK2021_allsites.csv'
csv_file2 = r'F:\wenqu\DOE_Arctic_project\2021_statistic\plant_cover.csv'

# Read the CSV files into DataFrames
df1 = pd.read_csv(csv_file1)
df2 = pd.read_csv(csv_file2)

# Merge the DataFrames on 'site', 'quadrat', 'spp'
merged_df = pd.merge(df1, df2, on=['site', 'quadrat', 'spp'])

# Optional: Save the merged DataFrame to a new CSV file
merged_csv_file = r'F:\wenqu\DOE_Arctic_project\2021_statistic\merged_csv.csv'
merged_df.to_csv(merged_csv_file, index=False)
```

```python
# Assuming df is your existing DataFrame
df = pd.read_csv(r'F:\wenqu\DOE_Arctic_project\2021_statistic\merged_csv.csv') # If you're reading from a CSV file
# Assuming df is your DataFrame
df['site_quadrat'] = df['site'].astype(str) + '_' + df['quadrat'].astype(str)
df
```

```python
rows, columns = df.shape

print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
df.dropna(inplace=True)  # Drop rows with NaN
rows, columns = df.shape

print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
print(df.columns)
```

```python
column_data_type = df['sla '].dtype
column_data_type
```

```python
df['sla '] = df['sla '].astype(float)
```

```python
# Compute the sum of 'cover' for each group and assign it to a new column
df['cover_sum'] = df.groupby(['site', 'quadrat'])['cover'].transform('sum')
df
```

```python
df['cover_ratio'] = df['cover'] / df['cover_sum']
df
```

```python
df['sla_cover'] = df['sla '] * df['cover_ratio']
df['la_cover'] = df['la'] * df['cover_ratio']
df['LDMC_cover'] = df['LDMC'] * df['cover_ratio']
df.drop(['site', 'quadrat'], axis=1, inplace=True)
df
```

```python
grouped_sla_cover_sum = df.groupby('site_quadrat')[['sla_cover', 'la_cover', 'LDMC_cover']].sum().reset_index()

# Rename the summed column for clarity
grouped_sla_cover_sum.rename(columns={'sla_cover': 'summed_sla_cover'}, inplace=True)

# Display the result
print(grouped_sla_cover_sum)
```

```python
# Assuming df is your DataFrame
grouped_sla_cover_sum.to_csv(r'F:\wenqu\DOE_Arctic_project\2021_statistic\grouped_sla_la_ldmc.csv', index=False)
```
