```python
import pandas as pd
```

```python
# read specific columns of csv file using Pandas
df_plant_cover = pd.read_csv(r"F:\wenqu\DOE_Arctic_project\Cover_and_chemistry_union_2021_2022.csv", usecols = ['site_quadrant', 'site_quadrant_species', 'cov1', 'pC', 'd13C', 'pN', 'd15N']).dropna()
df_plant_cover
```

```python
df_averages = df_plant_cover.groupby('site_quadrant_species', as_index=False).agg({'pC': 'mean', 'd13C': 'mean', 'pN': 'mean', 'd15N': 'mean', 'cov1': 'mean', 'site_quadrant': 'first'})
print(df_averages)
```

```python
# Count the number of unique items in 'Column'
unique_count = df_averages['site_quadrant'].nunique()

print("Number of unique items:", unique_count)
```

```python
df1 = df_averages.groupby(['site_quadrant'], as_index=False)['cov1'].sum()
df1.rename(columns={'cov1': 'sum_cover'}, inplace=True)
# df1["site_quadrant"] = df1['Site'].astype(str) + '_' + df1['quadrat'].astype(int).astype(str)
df1
```

```python
df_combine = pd.merge(df_averages, df1, on ='site_quadrant')
df_combine
```

```python
# df = df_combine[['site_quadrant', 'site_quadrant_species', 'sla', 'la', 'LDMC', 'cover', 'sum_cover']]

df_combine['percent_cover'] = df_combine['cov1'] / df_combine['sum_cover']
df_combine
```

```python
df_combine['weighted_pc'] = df_combine['pC'].astype(float) * df_combine['percent_cover'].astype(float)
df_combine['weighted_d13c'] = df_combine['d13C'].astype(float) * df_combine['percent_cover'].astype(float)
df_combine['weighted_pn'] = df_combine['pN'].astype(float) * df_combine['percent_cover'].astype(float)
df_combine['weighted_d15n'] = df_combine['d15N'].astype(float) * df_combine['percent_cover'].astype(float)
df_combine
```

```python
df_final = df_combine.groupby(['site_quadrant']).agg({'weighted_pc':'sum','weighted_d13c':'sum','weighted_pn':'sum', 'weighted_d15n':'sum'})
df_final
```

```python
df_final.to_csv(r'F:\wenqu\DOE_Arctic_project\cover_and_chemistry.csv') 
```
