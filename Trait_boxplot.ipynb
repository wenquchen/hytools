# community weighted

```python
leaf_cover_df = pd.read_csv(r'E:\wenqu\DOE_Arctic_project\2021_data\d15n_d13c_pc_pn.csv')
```

```python
leaf_cover_df
```

```python
leaf_cover_df['covre_Sum'] = leaf_cover_df.groupby(['site', 'quadrat'])['abundance_per'].transform('sum')
```

```python
leaf_cover_df
```

```python
leaf_cover_df['cover_percent'] = leaf_cover_df['abundance_per'] / leaf_cover_df['covre_Sum']
```

```python
leaf_cover_df
```

```python
leaf_cover_df['weighted_ldmc'] = leaf_cover_df['ldmc'] * leaf_cover_df['cover_percent']
leaf_cover_df['weighted_la'] = leaf_cover_df['la'] * leaf_cover_df['cover_percent']
leaf_cover_df['weighted_sla'] = leaf_cover_df['sla'] * leaf_cover_df['cover_percent']
# leaf_cover_df['weighted_d15N'] = leaf_cover_df['d15N'] * leaf_cover_df['cover_percent']
```

```python
leaf_cover_df
```

```python
leaf_cover_df.to_csv(r'E:\wenqu\DOE_Arctic_project\PFTS\box_plot_trait\site1b_leaf-weighted_traits.csv', index=False)
```

# Boxplot

```python
df = pd.read_csv(r'E:\wenqu\DOE_Arctic_project\PFTS\box_plot_trait\leaf_trait_PFT.csv')
```

```python
df
```

```python
# Get the minimum value of 'Column1'
min_value = df['weighted_sla'].min()

# Get the maximum value of 'Column1'
max_value = df['weighted_sla'].max()

print(f"Range of Column1: {min_value} to {max_value}")
```

```python
# Create a box plot with 'pft' as x-axis and 'sla' as y-axis
plt.figure(figsize=(10, 6))
sns.boxplot(x='pft', y='weighted_sla', data=df,fill=False, gap=.1)

# Add title and labels
plt.title('Box Plot of SLA by PFT')
plt.xlabel('PFT')
plt.ylabel('SLA')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pft', y='weighted_LDMC', data=df,fill=False, gap=.1)

# Add title and labels
plt.title('Box Plot of LDMC by PFT')
plt.xlabel('PFT')
plt.ylabel('LDMC')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pft', y='weighted_la', data=df,  fill=False, gap=.1)

# Add title and labels
plt.title('Box Plot of LA by PFT')
plt.xlabel('PFT')
plt.ylabel('LA')

# Show the plot
plt.show()
```
