```python
import pandas as pd
import numpy as np
```

```python
# read specific columns of csv file using Pandas
df_height = pd.read_csv(r"F:\wenqu\DOE_Arctic_project\2022_data\Summer2022_abundance.csv", usecols = ['Site','Quadrat', 'species', 'heighta', 'heightb', 'heightc '])
df_height
```

```python
df_height = df_height.replace(np.nan, 0)
df_height
```

```python
heighta = np.array(df_height['heighta'])
heightb = np.array(df_height['heightb'])
heightc = np.array(df_height['heightc '])
```

```python
heighta[0]
```

```python
height_abc = np.where((heighta > 0) & (heightb > 0) & (heightc > 0))
height_ab = np.where((heighta > 0) & (heightb > 0) & (heightc == 0))
height_a = np.where((heighta > 0) & (heightb == 0) & (heightc == 0))
height_o = np.where((heighta == 0) & (heightb == 0) & (heightc == 0))
```

```python
len(height_abc[0]), len(height_ab[0]), len(height_a[0]), len(height_o[0])
```

```python
heighta.shape[0]
```

```python
height_mean = []
for i in range(heighta.shape[0]): 
#     print(i)
    
    if i in height_abc[0]:
        print(i)
        mean_h = (heighta[i] + heightb[i] + heightc[i])/3
    
    if i in height_ab[0]:
        print(i)
        mean_h = (heighta[i] + heightb[i])/2
        
    if i in height_a[0]:
        print(i)
        mean_h = heighta[i]
    if i in height_o[0]:
        print(i)
        mean_h = 0
    height_mean.append(mean_h)
    

```

```python
height_mean
```

```python
df_height['meann_height'] = height_mean
df_height
```

```python
df_sum = df_height.groupby(['Site','Quadrat'], as_index=False)['meann_height'].sum()
df_sum
```

```python
df_sum.to_csv(r'F:\wenqu\DOE_Arctic_project\2022_data_statistics\site1b_height.csv')
```
