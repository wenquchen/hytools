```python
import matplotlib.pyplot as plt
import numpy as np
la_trait_pdf_ymax = np.load('F:/wenqu/Aviris/combine/la_trait_pdf_ymax.npy',  allow_pickle=True)

# Create a histogram with 5 bins
plt.hist(la_trait_pdf_ymax[:, 0].astype(np.float64), bins=50)

# Add labels and a title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('y_max data')

# Display the plot
plt.show()
```

```python
from scipy import stats
```

```python
la_trait_pdf_ymax_boxcox, _ = stats.boxcox(la_trait_pdf_ymax[:, 0].astype(np.float64))
```

```python
# Create a histogram with 5 bins
plt.hist(la_trait_pdf_ymax_boxcox, bins=50)

# Add labels and a title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('boxcox transformation y_max data')

# Display the plot
plt.show()
```
